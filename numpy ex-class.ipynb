### VECTORISATION

```python
def sourabh(a):
    
    if a%2==0:
        a +=5
    else:
        a -=5
    return a        
```

```python
a = np.array([1,2,3,4])
a
```

```python
# normal function can't be run on arrays
sourabh(a)
```

```python
# to make normal function run on array, we can do it in two ways
# 1. list comprehension
# 2. vectorisation
```

```python
# 1
[sourabh(i) for i in a]
```

```python
# 2
sourabh_vect = np.vectorize(sourabh)
sourabh_vect(a)
```

### Image Manipulation

```python
import matplotlib.pyplot as plt
```

```python
jacky = np.array(plt.imread('dog.jpg'))
plt.imshow(jacky)
```

```python
jacky
```

```python
# in numpy - 1600- matrices, 2560-rows, 3-columns
# in matplotlib - 1600- matrices, 2560-rows, 3-channels
jacky.shape
```

```python
jacky.ndim
```

```python
jacky_copy = jacky.copy()
```

```python
jacky_copy
```

```python
0-red channel
1-green channel
2-green channel
```

```python
# converting to all red
jacky_copy[:,:,(1,2)]=0
```

```python
plt.imshow(jacky_copy)
```

```python
jacky_copy[:,:,(0,2)]=1
```

```python
plt.imshow(jacky_copy)
```

### Cropping image

```python
# jacky[350:700  (ROWS),1200:1650  (COLUMNS),:   (CHANNELS)]
plt.imshow(jacky[350:700,1200:1650,:])
```

### BLURRING IMAGE

```python
# jacky[::2,::2,:]  selecting all rows and columns with step size of 2(skipping one pixels each)
# axis are changed with respect to original image as image gets lowered in quality  
plt.imshow(jacky[::2,::2,:])
```

```python
plt.imshow(jacky[::10,::10,:])
```

### Making Mirror Image

```python
# making only column reverse
plt.imshow(jacky[::,::-1,:])
```
